```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
import os

train_path = '/home/emmaykoushal/Documents/ENGINEERING/SEM6/MIMIA/JCOMP/Covid19-dataset-HistEq/train'
test_path = '/home/emmaykoushal/Documents/ENGINEERING/SEM6/MIMIA/JCOMP/Covid19-dataset-HistEq/test'
vectorized = np.array([[0, 0, 0]])

for folder in os.listdir(test_path):
  sub_path = test_path + '/' + folder
  for img in os.listdir(sub_path):
    image_path = sub_path + '/' + img
    image = cv2.imread(image_path)
    image = cv2.resize(image, (1000, 1000))
    vect = image.reshape((-1, 3))
    vectorized = np.concatenate((vectorized, vect), axis=0)
  
# for folder in os.listdir(test_path):
#   sub_path = test_path + '/' + folder
#   for img in os.listdir(sub_path):
#     image_path = sub_path + '/' + img
#     image = cv2.imread(image_path)
#     image = cv2.resize(image, (1000, 1000))
#     vect = image.reshape((-1, 3))
#     vectorized = np.concatenate((vectorized, vect), axis=0)

print(vectorized.shape)
```

```python
kmeans = KMeans(n_clusters=5, random_state = 0, n_init=5).fit(vectorized)

centers = np.uint8(kmeans.cluster_centers_)
```

```python
vect1 = test_img1.reshape((-1, 3))
pred1 = kmeans.predict(vect1)
pred1
```
